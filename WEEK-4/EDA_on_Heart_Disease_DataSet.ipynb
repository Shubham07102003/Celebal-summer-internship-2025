```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df=pd.read_csv('/content/heart-disease.csv')
df.head(10)
# Inference:
#The dataset contains patient records with features such as age, sex, chest pain type, cholesterol, blood pressure, and heart disease outcome (target).")
```

```python
#Data Information and Summary Statistics
df.info()
'''
Inference:
The dataset has 303 entries and 14 features.
All columns are complete with no missing values.
Key statistics (mean, std, min, max) for each feature are shown for quick reference.
'''
```

```python
df.describe()
```

```python
#Missing Values Check
df.isna().sum()

'''
Inference:
There are no missing values in any column, ensuring the dataset is complete for analysis.'''
```

```python
df['age'].plot.hist(bins=10);

#  5.1 Age Distribution
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()


Inference:
Most patients are between 40 and 70 years old, with a peak around the mid-50s.
```

```python
# 5.2 Sex Distribution

sex_counts = df['sex'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(sex_counts, labels=['Female', 'Male'], autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
plt.title('Sex Distribution')
plt.axis('equal')
plt.show()


'''
Inference:
The dataset is male-dominated, with males comprising about 68% of the patients.'''
```

```python
#5.3 Boxplots for Outlier Detection

plt.figure(figsize=(15,10))
for i, col in enumerate(['age', 'trestbps', 'chol', 'thalach', 'oldpeak']):
    plt.subplot(2, 3, i+1)
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()


'''
Inference:
Outliers are visible in cholesterol (chol), resting blood pressure (trestbps), and oldpeak.
These may represent rare but clinically significant cases or data entry errors. '''
```

```python
over_50 = df[df["age"] > 50]
over_50.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);
```

```python
# 6.1 Correlation Heatmap

plt.figure(figsize=(12,8))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()


'''
Inference:
Age and resting blood pressure (trestbps) are positively correlated.
Max heart rate (thalach) is negatively correlated with heart disease (target).
Oldpeak also shows a negative correlation with target, suggesting higher values may be associated with disease. '''
```

```python
# 6.2 Age vs Cholesterol by Heart Disease Status

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='age', y='chol', hue='target', palette='coolwarm')
plt.title('Age vs Cholesterol by Heart Disease Target')
plt.show()


'''
Inference:
Higher cholesterol levels are seen across all ages, but patients with heart disease (target=1) are spread across the age and cholesterol spectrum. '''
```

```python
# 6.3 Age vs Max Heart Rate by Heart Disease Status

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='age', y='thalach', hue='target', palette='coolwarm')
plt.title('Age vs Max Heart Rate by Heart Disease Target')
plt.show()


'''
Inference:
Patients with heart disease tend to have lower maximum heart rates (thalach), especially as age increases. '''
```

```python
fig,ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol")

# Add a legend
ax.legend(*scatter.legend_elements(),title="Target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle='--');
```

```python
# Set the style
plt.style.use('seaborn-whitegrid')

# OO method from scratch
fig,ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"],
                   cmap="winter") # This changes the color scheme

# Customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol")

# Add a legend
ax.legend(*scatter.legend_elements(),title="Target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle='--');
```

```python
# Customizing the x and y axis limitations

plt.style.use("ggplot")

# Subplot of chol, age, thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,
                        ncols=1,
                        figsize=(10,10),sharex=True)

# Add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"],
                   cmap="winter")

# Customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol")

# Change the x limitation
ax0.set_xlim([50,80])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target")

# Add a horizontal line to ax0
ax0.axhline(over_50["chol"].mean(),linestyle='-.');

# Add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"],
                   cmap="winter")

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate Levels",
       xlabel="Age",
       ylabel="Max Heart Rate");

# Change the x limitation
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target")

# Add a horizontal line to ax1
ax1.axhline(over_50["thalach"].mean(),linestyle='-.')

# Add a title toi the entire figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight='bold');
```
